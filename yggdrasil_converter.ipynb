```python
import os
from pathlib import Path
from matplotlib import pyplot as plt
from LWphotorates.stellar_spectra.utils import convert_yggdrasil_spec_to_xr, DATA_DIR

plt.style.use(Path(os.environ["HOME"]) / "default.mplstyle")
```

```python
in_file_path = DATA_DIR / "Yggdrasil_PopIII_TH.txt"
out_file_path = in_file_path.parent / in_file_path.name.replace(".txt", ".nc")
ds_description = "Yggdrasil PopIII.1 (extremely top-heavy IMF) spectra. For more info look here: https://www.astro.uu.se/~ez/yggdrasil/ReadMe.txt."
ygg_th_ds = convert_yggdrasil_spec_to_xr(in_file_path, ds_description, out_file_path)

in_file_path = DATA_DIR / "Yggdrasil_PopIII_BH.txt"
out_file_path = in_file_path.parent / in_file_path.name.replace(".txt", ".nc")
ds_description = "Yggdrasil PopIII.2 (moderately top-heavy IMF) spectra. For more info look here: https://www.astro.uu.se/~ez/yggdrasil/ReadMe.txt."
ygg_bh_ds = convert_yggdrasil_spec_to_xr(in_file_path, ds_description, out_file_path)

in_file_path = DATA_DIR / "Yggdrasil_PopIII_Kr.txt"
out_file_path = in_file_path.parent / in_file_path.name.replace(".txt", ".nc")
ds_description = "Yggdrasil PopIII (Kroupa IMF) spectra. For more info look here: https://www.astro.uu.se/~ez/yggdrasil/ReadMe.txt."
ygg_kr_ds = convert_yggdrasil_spec_to_xr(in_file_path, ds_description, out_file_path)
```

```python
fig, ax = plt.subplots()
ax.plot(ygg_th_ds.wavelength, ygg_th_ds.luminosity.isel(age=0))
ax.plot(ygg_bh_ds.wavelength, ygg_bh_ds.luminosity.isel(age=0))
ax.plot(ygg_kr_ds.wavelength, ygg_kr_ds.luminosity.isel(age=0))
ax.set_yscale("log")
ax.set_xscale("log")
```
